```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Import "numpy" could not be resolved
from sklearn import linear_model, metrics # Import "sklearn" # Đọc dữ liệu từ file csv
df = pd.read_csv("ds_n_v2.csv", index_col = 0)
print(df)
# Thêm cột mới, giới tỉnh Nam = 1, giới tỉnh Nữ = 0
df[ 'GT'] = df. GioiTinh.apply(lambda x: 1 if x== 'Nam' else 0)
print(df)
```

```
# sử dụng hồi quy tuyến tính
X = df.loc[:, ['ChieuCao', 'GT']].values
y = df.CanNang.values
model = linear_model.LinearRegression()
model.fit(X, y)
```

```
mse = metrics.mean_squared_error(model.predict(X), y)
r_2 = metrics.r2_score(y, model.predict(X))
print("Hệ số xác định R2:", r_2)
print( "Tổng bình phương sai số trên tập mẫu:", mse)
print ("Hệ số hồi quy:", model.coef_)
print ("Sai số:", model.intercept_)
print(f"Công thức: [Nặng] = {model.coef_} x [Cao, Giới tính] + {model.intercept_}")
```

```
# dự báo một số tình huống
while True:
    height_input = input("Nhập chiều cao (nhập 0 để dừng): ")
    try:
        x = float(height_input.replace("cm", "").strip()) # Remove "cm" and leading/trailing spaces before conversion
    except ValueError:
        print("Vui lòng nhập chiều cao là số.")
        continue
    if x <= 0:
        break
    print("Nam giới cao", x, "cm, dự báo cân nặng", model.predict([[x, 1]])) # Use model.predict with the correct input format
    print("Nữ giới cao", x, "cm, dự báo cân nặng", model.predict([[x, 0]])) # Use model.predict with the correct input format
```
